```python
# finding out the season based on the year and creating a new column
matches['season'] = pd.DatetimeIndex(matches['date']).year
matches.head(2)
```

```python
runs_data_frame = [] #creating a empty list
print('number of matches played in each season')
# taking the unique value of season from the column
for i in matches['season'].unique(): # unique values will give the number of seasons played 
    # printing the number of matches played in each season 
    runs_data_frame.append({'season' : i , 'No_of_Matches' : len(matches[matches['season'] == i]) })
    
matches_per_season_df = pd.DataFrame(runs_data_frame)

# Display the created dataframe
matches_per_season_df.head()
```

```python
# merging two datasets with required columns and creating a new dataset
data = matches[['season','id']].merge(deleviries , left_on= 'id' , right_on='id' , how='left').drop('id', axis=1)
```

```python
# visualization of total runs in each season
colors = ['skyblue',] * 13
colors[5] = 'darkmagenta'
#providing the dataset and the values of x and y
fig = px.bar( data_frame=runs_per_season ,x = runs_per_season.season,y = runs_per_season.total_runs)
#providing the title for the visualization
fig.update_layout(title= "Runs Per Season")
#proving the marker line width , colors etc.
fig.update_traces(marker_line_color= 'Black' , marker_line_width=1.2 , 
                  marker_color=colors)
```

```python
#creating a funtion to find out the lucky vanue for particular team
def lucky(match_data, team_name, top_n=10):
    #getting the reult of each team on each vanue
    result_series = match_data[match_data['winner'] == team_name]['venue'].value_counts().nlargest(top_n)
    #converting to the dataframe and and sorting
    result_df = pd.DataFrame({'Venue': result_series.index, 'Wins': result_series.values})
    return result_df
```

```python
# no of wins of each team on each vanue
kkr_data = lucky(matches , 'Kolkata Knight Riders')
mi_data = lucky(matches , 'Mumbai Indians')
csk_data = lucky(matches , 'Chennai Super Kings')
rcb_data = lucky(matches , 'Royal Challengers Bangalore')
kxip_data = lucky(matches , 'Kings XI Punjab')
rr_data = lucky(matches , 'Rajasthan Royals')
dd_data = lucky(matches , 'Delhi Daredevils')
Srh_data = lucky(matches , ' Sunrisers Hyderabad')
dch_data = lucky(matches , 'Deccan Chargers')
dc_data = lucky(matches , 'Delhi Capitals')
gl_data = lucky(matches , 'Gujarat Lions')
pw_data = lucky(matches , 'Pune Warriors')
rps_data = lucky(matches , 'Rising Pune Supergiant')
ktk_data = lucky(matches , ' Kochi Tuskers Kerala')
    
```
