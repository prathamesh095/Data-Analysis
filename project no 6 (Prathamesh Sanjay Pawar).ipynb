```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/AshishJangra27/Data-Science-Specialization/main/Week%20-%208/Projects/country_vaccinations.csv')
```

```python
df.head()
```

```python
df['date']            # printing the date column
```

```python
# creating a empty list and appending the data of date column into the list and also slicing the data

month = []         # creating a empty list

for i in df['date']:
    month.append(i[5:7]) # appending and slicing of data
```

```python
# creating a dictonary which contains months with their month number

dct = {'01': "Jan",'02': "Feb",'03': "Mar",'04': "Apr",
       '05': "May",'06': "Jun",'07': "Jul",'08': "Aug",
       '09': "Sept",'10': "Oct",'11': "Nov",'12': "Dec", }
```

```python
for i in dct.keys():    # printing the keys
    print(i)
```

```python
month
```

```python
# creating a list and aapending the data of month and month number by which i have created column

month_1 = []         # creating a empty list

for num in month:
    for i in dct.keys():         # reading the keys of dictonaries
        if (num == i):           # checking whether num in month is equal to keys in dct or not          
           month_1.append(dct[i])  # if it is equal then appending the relevant data to a empty list
```

```python
df['Months'] = month_1          # creating a Months column
df['Month_num'] = month         # creating a Month column
```
