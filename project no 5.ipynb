```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/AshishJangra27/Data-Science-Specialization/main/Week%20-%208/Projects/country_vaccinations.csv')
data.head()
```

 ## Q1) Adding year column form date column

```python
year = []                ## creating a empty list
 
for i in data['date']:
     year.append(i[:4])            ## adding the details in empty list using slicing
print(year)                        ## printing the final list
```

```python
 ## creating a new column and adding into the original dataset
data['Year'] = year  
```

```python
data.head()
```

```python
## it is the year coulmn which i have added from data column
data['Year']      
```

## Q2) Adding month column form date column 

```python
month = []             ## creating a empty list 

for i in data['date']:
    month.append(i[5:7])    ## adding the month column from data column using slicing
print(month)                ## printing the final list
```

```python
## creating a month coulmn in original dataset
data['Month'] = month
```

```python
data.head()
```

```python
## it sa the month column which we have added into a original dataset from data coulumn
data['Month']
```

## Q3) Creating a day column from date column  

```python
day = []            ## creating a empty list

for i in data['date']:          
    day.append(i[8:])              ## adding the datails in empty list using slicing
print(day)                         ## printing the final list
```

```python
## Creating a day column and adding into the original dataset

data['Day'] = day
```
